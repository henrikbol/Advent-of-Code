```python
# data = p.examples[0].input_data
data = p.input_data

pairs_str = [i.strip().split("-") for i in data.split(",")]
pairs = [(int(start), int(end)) for start, end in pairs_str]
```

```python
def is_valid(s: str) -> bool:
    n = len(s)
    if n % 2 != 0:
        return False
    m = n // 2
    return s[0:m] == s[m : n]
```

```python
def has_repeating_pattern(s: str) -> bool:
    n = len(s)
    for i in range(1, n // 2 + 1):   
        if n % i == 0:    
            pattern = s[:i]
            if pattern * (n // i) == s:
                return True

    return False
```

```python
has_repeating_pattern("123123123123"), has_repeating_pattern("1212"), has_repeating_pattern("1234"), has_repeating_pattern("1111112")
```

```python
# Part 1
%time
res = 0

for start, end in pairs:
    for i in range(start, end + 1):
        if is_valid(str(i)):
            res += i

res
```

```python
# Part 2
%time
res = 0

for start, end in pairs:
    for i in range(start, end + 1):
        if has_repeating_pattern(str(i)):
            res += i

res
```
