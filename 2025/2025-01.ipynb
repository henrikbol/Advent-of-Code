```python
def get_directions(d: str) -> list[int]:
    return [
        -int(num) if turn == "L" else int(num)
        for turn, num in re.findall(r"([LR])(\d+)", d)
    ]


def get_new_pos(pos: int, step: int) -> (int, int):
    pass_0 = 0
    if (pos == 0) and (step < 0):
        # Edge case if you are at pos 0 and move backwards. Do not count the first rotation
        pass_0 = -1
    pos += step
    while pos < 0:
        pos += 100
        pass_0 += 1
    while pos > 99:
        pos -= 100
        if pos != 0:
            # Another edge case. It you move forward and end at 0. Do not count this one as it will also satisfy the "pos==0" check in the main loop.
            pass_0 += 1
    return pos, pass_0
```

```python
%%time
# Part A
pw = 0
pos = 50
input_data = p.input_data
# input_data = p.examples[0].input_data
for step in get_directions(input_data):
    pos, _ = get_new_pos(pos, step)
    if pos == 0:
        pw += 1

print(pw)
```

```python
%%time
# Part B
pw = 0
pos = 50
input_data = p.input_data
# input_data = p.examples[0].input_data
l = []
for step in get_directions(input_data):
    pre = f"Pos: {pos} Step: {step}"  # For debugging
    pos, pass_0 = get_new_pos(pos, step)
    pw += pass_0
    if pos == 0:
        pw += 1

    l.append(f"{pre} Ny: {pos}, Pass: {pass_0}, PW: {pw}")  # For debugging

print(pw)
```
